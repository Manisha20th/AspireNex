```python
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
df=pd.read_csv('titanic.csv')
```

```python
df.head()
```

```python
df.shape
```

#droping Unwanted Features

```python
df.drop(columns=['PassengerId','Name','Ticket','Cabin'],inplace=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

#handling missing values in Age,Fare

```python
mean_age = df['Age'].mean()
df['Age'].fillna(mean_age, inplace=True)
mean_fare = df['Fare'].mean()
df['Fare'].fillna(mean_fare, inplace=True)
```

```python
df['Age'].isnull().sum()
```

```python
df['Fare'].isnull().sum()
```

```python
df
```

```python
df['Sex'] = df['Sex'].map({'male' : 0, 'female' : 1})
df['Embarked'] = df['Embarked'].map({'Q' : 0, 'S' : 1, 'C':2})
```

```python
X_train,X_test,y_train,y_test=train_test_split(df.drop(columns=['Survived']),df['Survived'],test_size=0.3,random_state=42)
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_train_pred=model.predict(X_train)
```

```python
y_train_pred
```

```python
T_a = accuracy_score(y_train, y_train_pred)
T_a
```

```python
Y_pred_test=model.predict(X_test)
```

```python
Test_ac=accuracy_score(y_test,Y_pred_test)
Test_ac
```
